```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
from matplotlib.colors import colorConverter
```

```python
xp = 0.0
yp = 1.3
rp = 0.5

kappa_star = np.arccos((1 + yp * yp - rp * rp) / (2 * yp))
kappa = np.arccos((rp * rp + yp * yp - 1) / (2 * rp * yp))
x1 = np.sin(kappa_star)
y1 = np.cos(kappa_star)
x2 = - x1
y2 = y1


star = plt.Circle((0, 0), 1, color='#ffae36', alpha=0.75, ec='k')
    
planet = plt.Circle((xp, yp), rp, color=colorConverter.to_rgba('r', alpha=0.0), ec='k')

plt.figure(figsize=(10, 10))
plt.ylim(-0.1, 1.9)
plt.xlim(-1, 1)
ax = plt.gca()

ax.add_patch(star)
ax.add_patch(planet)

plt.plot(0, 0, 'ko')
plt.plot(xp, yp, 'ko')
plt.plot((0.0, x1), (0.0, y1), color='k')
plt.plot((0.0, x2), (0.0, y2), color='k')

plt.plot((xp, x1), (yp, y1), color='k')
plt.plot((xp, x2), (yp, y2), color='k')
plt.plot((0, 0), (0, yp), 'k--')

star_arc = Arc((0, 0), 0.5, 0.5, theta2 = 90, theta1 = 90 - 180 * kappa_star / np.pi)
planet_arc = Arc((xp, yp), 0.3, 0.3, theta1 = -90, theta2 = - (90 - 180 * kappa / np.pi))
ax.add_patch(star_arc)
ax.add_patch(planet_arc)

plt.annotate(r"$\kappa$", (0.03, 1.1), fontsize=20)
plt.annotate(r"$\kappa^*$", (0.03, 0.3), fontsize=20)
plt.annotate(r"$r$", (-0.3, 1.1), fontsize=20)
plt.annotate(r"$b$", (-0.08, 0.6), fontsize=20)
plt.annotate(r"$1$", (-0.3, 0.55), fontsize=20)

ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
plt.tight_layout()
plt.savefig("/Users/tgordon/Desktop/star_overlap.pdf")
```
